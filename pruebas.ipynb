```python
#4.Crea un vector de 10 elementos, siendo los índices impares unos y los índices pares dos.
x = np.zeros((1,10))
for i in range(len(x[0])):
    if i%2 == 0:
        x[0,i]= 2
       
    else:
        x[0,i]=1

print(x)
```

```python
x = np.zeros((1, 10))  # Inicializar todos los valores en cero

for i in range(len(x[0])):  # Solo necesitas iterar sobre la dimensión interna de x
    if i % 2 == 0:
        x[0, i] = 2

    else:
        x[0, i] = 1

print(x)
```

```python
import matplotlib.pyplot as plt
#ax = plt.axes() #Se crea un objeto axes
```

```python
x = np.radians(np.linspace(0,360))
x1 = np.sin(x)
x2 = np.cos(x)

plt.subplot(2,1,1)
plt.ylim(-1,1)
plt.xlim(0, 6)
plt.xlabel("Ángulo(rad)")
plt.ylabel("Sin(x)")
plt.title("Función seno")
plt.plot(x,x1, "r--", label = "Seno")


plt.subplot(2,1,2)
plt.ylim(-1,1)
plt.xlim(0, 6)
plt.xlabel("Ángulo(rad)")
plt.plot(x,x1, "r--", label ="Seno" )
plt.plot(x,x2, "g-", label ="Coseno" )
plt.legend()

plt.subplots_adjust(hspace=0.4, wspace=1)
plt.show()
```
