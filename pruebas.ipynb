```python
import matplotlib.pyplot as plt
#ax = plt.axes() #Se crea un objeto axes
```

```python
x = np.radians(np.linspace(0,360))
x1 = np.sin(x)
x2 = np.cos(x)

plt.subplot(1,2,1)
plt.ylim(-1,1)
plt.xlim(0, 6)
plt.xlabel("Ángulo(rad)")
plt.ylabel("Sin(x)")
plt.title("Función seno")
plt.plot(x,x1, "r--", label = "Seno")


plt.subplot(1,2,2)
plt.ylim(-1,1)
plt.xlim(0, 6)
plt.xlabel("Ángulo(rad)")
plt.plot(x,x1, "r--", label ="Seno" )
plt.plot(x,x2, "g-", label ="Coseno" )
plt.legend()

plt.subplots_adjust(hspace=0.4, wspace=1)
plt.show()
```
